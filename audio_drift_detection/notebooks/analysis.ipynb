# Analysis Notebook

Load experiment results and create additional analysis plots.

```python
import json
from pathlib import Path
import pandas as pd

output_dir = Path('../outputs')

# Load metrics from all scenarios
results = []
for scenario in ['abrupt', 'gradual', 'noise']:
    mf = output_dir / scenario / 'metrics.json'
    if mf.exists():
        with open(mf) as f:
            results.append(json.load(f))

df = pd.DataFrame(results)
df
```

```python
# Load ablation results
abl_path = output_dir / 'ablation' / 'ablation_results.json'
if abl_path.exists():
    with open(abl_path) as f:
        abl = json.load(f)
    pd.DataFrame(abl)
```
